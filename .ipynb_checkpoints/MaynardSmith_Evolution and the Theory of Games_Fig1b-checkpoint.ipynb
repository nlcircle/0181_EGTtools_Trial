# EGT Example for two three strategies

In J. Maynard Smith's work 'Evolution and the Theory of Games', section 1 contains an example of the 'Hawk-Dove-Replicator' game. This is an extention of the more common 2-strategy 'Hawk-Dove' game, where an third phenotype is added.

This third phenotype or strategy behaves like a dove when engaging a dove and similarly, like a hawk when meeting a hawk. The payoff matrix for this 3-strategy game is as follows:

| Strategy | Hawk | Dove | Retaliator |
| :--- | :---: | :---: | ---: |
| Hawk | -1 | 2| -1 |
| Dove | 0 | 1 | 0.9 |
| Retaliator | -1 | 1.1 | 1 |

Solving for Evolutionary Stable Strategies (ESS) shows two solution, one with a mix between H and D and the second with only R:

| ESS | Hawk | Dove | Retaliator |
| :--- | --- | --- | ---: |
| I1 | 0.5 | 0.5 |  |
| I2 |  |  | 1 |

Results from Evolutionary Games can be represented in Simplex diagrams, using barymetric coordinates to reflect strategies and their gradients. We use EGTtools as a analysis tool for this particular game.

In J. Maynard Smith's book he captures this HDR game in table. 2, with the payoff table above as table 2.b. In the same section of the book, he presents fig. 1.b which represents the flows for the HDR game, showing attractors at I and R and a saddle point at S. It seems that the positioning of I in this figure on page 19 of the book is incorrect as it points to I = 1/3H+2/3D rather than I=1/2H+1/2D (our I2 from the table).

The purpose of this notebook is to recreate this figure with EGTtools, both to confirm a possible mistake in the book and as practice for setting up evolutionary game theoretic analyses and display options.

## Flow diagram (simplex) for HDR game

The code for displaying the HDR as per Maynard Smith's book is based on the following example in the EGTtools 'read the docs'.

```python
# get the necessary imports
import numpy as np
import matplotlib.pyplot as plt
from egttools.plotting.simplified import plot_replicator_dynamics_in_simplex


# Payoff matrix
#V = 2; D = 3; T = 1; F = 2; S = 1
A = np.array([
        [ -1, 2, -1], # Hawk
        [ 0, 1, 0.9], # Dove
        [ -1, 1.1, 1] # Replicator
    ])
type_labels = ['Hawk', 'Dove', 'Replicator']


# calculate the values and plot the figures
fig, ax = plt.subplots(figsize=(10,8))

simplex, gradient_function, roots, roots_xy, stability = plot_replicator_dynamics_in_simplex(A, ax=ax)

plot = (simplex.draw_triangle()
           .add_vertex_labels(type_labels, epsilon_bottom=0.1)
           .draw_stationary_points(roots_xy, stability)
           .draw_gradients(zorder=0)
           .add_colorbar()
           .draw_scatter_shadow(gradient_function, 100, color='gray', marker='.', s=0.1)
          )

ax.axis('off')
ax.set_aspect('equal')

plt.xlim((-.05,1.05))
plt.ylim((-.02, simplex.top_corner + 0.05))



```
